```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Blow_XY (Shape)

```python
BlowXY = pd.read_csv('Data/00-blow-XY-data.csv')
BlowXY
```

```python
idx = np.zeros(4, dtype=np.int64)
for ChartNo in range(2,6):
    idx[ChartNo-2] = np.where(BlowXY.iloc[:,0]=='Chart Line %d'%(ChartNo))[0][0]

idx
```

```python
BlowXY_1 = pd.DataFrame(BlowXY.iloc[        :idx[0]-3,:], dtype = 'float32')
BlowXY_2 = pd.DataFrame(BlowXY.iloc[idx[0]+4:idx[1]-3,:], dtype = 'float32')
BlowXY_3 = pd.DataFrame(BlowXY.iloc[idx[1]+4:idx[2]-3,:], dtype = 'float32')
BlowXY_4 = pd.DataFrame(BlowXY.iloc[idx[2]+4:idx[3]-3,:], dtype = 'float32')
BlowXY_5 = pd.DataFrame(BlowXY.iloc[idx[3]+4:        ,:], dtype = 'float32')

BlowXY_1.shape, BlowXY_2.shape, BlowXY_3.shape, BlowXY_4.shape, BlowXY_5.shape
```

```python
#plt.figure(figsize = (12,9))
plt.plot(BlowXY_1.iloc[:,0], BlowXY_1.iloc[:,1], 'o-')
plt.plot(BlowXY_2.iloc[:,0], BlowXY_2.iloc[:,1], 'o-')
plt.plot(BlowXY_3.iloc[:,0], BlowXY_3.iloc[:,1], 'o-')
plt.plot(BlowXY_4.iloc[:,0], BlowXY_4.iloc[:,1], 'o-')
plt.plot(BlowXY_5.iloc[:,0], BlowXY_5.iloc[:,1], 'o-')
plt.xlim([-0.25, 0])
plt.ylim([-0.25, 0])
plt.gca().invert_xaxis()
plt.xlabel('X [m]', fontsize=15)
plt.ylabel('Y [m]', fontsize=15)
plt.title('Blow - Shape(X,Y)', fontsize=20)
plt.legend(['Initial','Opt1','Opt2','Opt3','Opt4'])
plt.grid(alpha=0.5)
plt.show()
```
